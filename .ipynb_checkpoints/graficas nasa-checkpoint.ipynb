```python
import pandas as pd
import numpy as np
import calendar
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime
import re
pd.options.display.max_columns = None
```

```python
data = pd.read_csv("./nasa_exoplanets.csv")
data.head()
```

```python
data.shape , data.size
```

```python
data.columns
```

```python
new_columns = {i.replace("_"," " ) for i in data.columns}
data.columns = new_columns
data.sample(20)
```

```python
data.columns
```

```python
(data.isnull().sum()/data.shape[0])*100
```

```python
plt.figure(figsize = (8,5))
plt.hist(data["solar system name"], edgecolor = "black",
                              facecolor = "mediumseagreen",
                            bins = 5)
plt.show();
```

```python
nulos = (data.isnull().sum()/data.shape[0])*100
```

```python


plt.pie(nulos, 
       labels = data.columns);
plt.title("Porcentaje nulos") 
plt.legend(bbox_to_anchor=(1.2, 1))
plt.show();
```

```python
ax = plt.axes() 

ax.set_facecolor("white") 
ax.spines['bottom'].set_color('black') 
ax.spines['top'].set_color('white')
ax.spines['right'].set_color('white') 
ax.spines['left'].set_color('black') 
plt.plot(data["solar system name"], 
        data["stellar magnitude"], 
         color = "black", 
         linewidth = 2, 
         marker = "o")



plt.xlabel("planet type") 
plt.ylabel("planet temperature")
plt.xticks(data["solar system name"].unique(),  rotation=45) # modificar los nombres del eje X


plt.show();
```

```python
plt.scatter(data["solar system name"], data["planet density"])
plt.show()
```

```python
plt.scatter(data['solar system name'], data['stellar magnitude'])

plt.show(); 
```

```python
sns.barplot(x='solar system name', y='planet type',data=data);
```

```python
mask = np.triu(np.ones_like(data.corr(), dtype=np.bool))
mask
```

```python
sns.heatmap(data.corr(),
            cmap = "YlGnBu", 
            mask = mask, 
           annot = True); 
```

```python
sns.jointplot(data['planet mass'], data['solar system name']
             kind = 'hex', 
             color = 'seagreen' )
plt.show();
```

```python
sns.jointplot(data['planet mass'],
             kind = 'hex', 
             color = 'seagreen' )
plt.show();
```

```python
columnas = ["stellar magnitude", "planet orbital inclination", "host mass"]

for col in columnas: 
    plt.boxplot(data[col])
    
fig, ax = plt.subplots(1, 3) 
ax[0].boxplot(data['stellar magnitude']) 
ax[1].boxplot(data['planet orbital inclination']) 
ax[2].boxplot(data['host mass']) 
plt.show()
```
